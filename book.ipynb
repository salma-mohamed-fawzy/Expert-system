```python
################ book and novel recommendation #################

from experta import *

class findBook(KnowledgeEngine):
    @DefFacts()
    def initial(self):
        print("hello now you can find your favourite book \n let's try")
        yield Fact(action = "find_book")
    #######################################book or novel##########################################
    @Rule(NOT (Fact(book = W())))
    def value1(self):
        self.declare(Fact(book = input(" --> do you love books ? ")))
        
    @Rule(NOT (Fact(novel = W())))
    def value2(self):
        self.declare(Fact(novel = input(" --> do you love novels ? ")))
    ################################################################################
    @Rule( Fact(book = "yes"),NOT (Fact(sciencefiction = W())))
    def value3(self):
        self.declare(Fact(sciencefiction = input(" --> do you love sciencefiction books ? ")))
        
    @Rule( Fact(book = "yes"),NOT (Fact(programming = W())))
    def value4(self):
        self.declare(Fact(programming = input(" --> do you love programming langauges books ? ")))
    ##########################################################################################
    @Rule( Fact(book = "yes"), Fact(sciencefiction = "yes"))
    def value5(self):
        print("**** you may love harrypotter books ******")
          
    @Rule( Fact(book = "yes"), Fact(programming = "yes"), NOT(Fact(java = W())))
    def value6(self):
         self.declare(Fact(java = input(" --> do you want java programming langauge books ? ")))
   
    @Rule( Fact(book = "yes"), Fact(programming = "yes"), NOT(Fact(python = W())))
    def value7(self):
         self.declare(Fact(python = input(" --> do you want python programming langauge books ?")))
    ############################################################################################
    @Rule( Fact(book = "yes"), Fact(programming = "yes"), Fact(java = "yes"))
    def value8(self):
         print("**** you should study java hard way book ******")
         
    @Rule( Fact(book = "yes"), Fact(programming = "yes"), Fact(python = "yes"))
    def value9(self):
         print("**** you should study python hard way book ******")
            
    ###########################################################################################
    @Rule( Fact(novel = "yes"),NOT (Fact(romantic = W())))
    def value10(self):
        self.declare(Fact(romantic = input(" --> do you love romantic novel ? ")))
        
    @Rule( Fact(novel = "yes"),NOT (Fact(mystery = W())))
    def value11(self):
        self.declare(Fact(mystery = input(" --> do you love mystery  ? ")))
     ##########################################################################################
    @Rule( Fact(novel = "yes"), Fact(mystery = "yes"))
    def value12(self):
        print("**** you should read agatha cresty novels ******")
          
    @Rule( Fact(novel = "yes"), Fact(romantic = "yes"), NOT(Fact(old = W())))
    def value13(self):
         self.declare(Fact(old = input(" --> do you want old romantic novels ")))
         
    @Rule( Fact(novel = "yes"), Fact(romantic = "yes"), NOT(Fact(new = W())))
    def value14(self):
         self.declare(Fact(new = input(" --> do you want new romantic novels ")))
    ############################################################################################
    @Rule( Fact(novel = "yes"), Fact(romantic = "yes"), Fact(old = "yes"))
    def value15(self):
         print("**** you may love Titanic novel ******")
         
    @Rule( Fact(novel = "yes"), Fact(romantic = "yes"), Fact(new = "yes"))
    def value16(self):
         print("**** you may love partners ******")

engine = findBook()
engine.reset()
engine.run()
```
